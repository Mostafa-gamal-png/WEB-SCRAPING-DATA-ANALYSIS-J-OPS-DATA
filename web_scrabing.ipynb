```python
f=open('data_scrabed.csv','w')
header='jop_title,company_name\n'
f.write(header)
```

```python
for container in containers:
    jtitle=container.find_all('h2',{'class':'css-m604qf'})
    jop_title=jtitle[0].text.strip()
    
    cname=container.findAll('a',{'class':"css-17s97q8"})
    company_name=cname[0].text.strip()
    #print(jop_title)
    #print(company_name)
    #print()
    f.write(jop_title+','+company_name+'\n')
f.close()    
```
